```python
import pandas as pd
arquivo = pd.read_csv('wine_dataset.csv')
```

```python
arquivo.head()
```

```python
arquivo.tail()
```

```python
# classificando em vinho red = 0, e white = 1:

# Mofificando a coluna style: red p/ 0, e white p/ 1
arquivo['style'] = arquivo['style'].replace('red', 0)
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
arquivo.head()
```

```python
arquivo.tail()
```

```python
# Separando as variáveis entre preditoras e variável alvo:

# alvo variável y:
y = arquivo['style']

x = arquivo.drop('style', axis = 1) # sem a coluna style
```

```python
y
```

```python
x
```

```python
# Importando a função do sklearn: train_teste_split:
from sklearn.model_selection import train_test_split

# Criando os conjuntos de dados de treino e teste:
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) # 30% p treino
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
x_teste.shape
```

```python
y_teste.shape
```

```python
# Machine Learning ------------------------------------------------------------:

# Importando a função do sklearn -> ExtraTreesClassifier:
from sklearn.ensemble import ExtraTreesClassifier

# Criação do modelo:
modelo = ExtraTreesClassifier(n_estimators=100) # n_estimators sem estimadores
modelo.fit(x_treino, y_treino) # Aplicar o algoritmo aos dados - Treino 

# Imprimindo resultados:
resultado = modelo.score(x_teste, y_teste) # Comparar dados de x_teste com y_teste
print("Acurácia", resultado)
```

```python
# Filtrando algumas amostras:
y_teste[400:403] # da 400 até 403
```
